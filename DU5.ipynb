## Sparse matice (sousednosti)
Vytvořte sparse matici o rozměrech $n^2\times n^2$. Tato matice reprezentuje pravidelnou čtvercovou síť o velikosti $n \times n$. Každá buňka této čtvercové sítě odpovídá jednomu řádku/sloupci matice, přičemž pořadí buněk čtvercové sítě berte po řádcích (viz obrázek níže).

Hodnoty v matici budou řízeny jedním parametrem $\varepsilon$; pro jednoduchost si místo něj představte hodnotu $0.1$.
Nenulové hodnoty v matici jsou zadány sousedností buněk následovně:
- hodnoty na diagonále (tedy sousednost prvku se sebou samým) jsou $1 + k \cdot \varepsilon$ (kde $k$ určuje počet sousedních buněk: 4 uvnitř, 3 na hranách, 2 v rozích),
- mimodiagonální prvky mají hodnotu $-\varepsilon$ právě tehdy, pokud index řádku a index sloupce mimodiagonálního prvku odpovídá sousedním buňkám ve čtvercové síti (viz obrázek).

**Ukázka, jak přechází mřížka do matice pro $n=4$ a $\varepsilon = 0.1$, je na následujícím obrázku:**


Vznikne matice s následujícími vlastnostmi:
- velikost $n^2 \times n^2$,
- $\varepsilon$ je parametr, který můžete nastavit, například $\varepsilon = 0.1$,
- matice je symetrická, tj. $A = A^T$,
- každý řádek/sloupec má maximálně 5 nenulových prvků,
    - 5, pokud se jedná o buňku uvnitř sítě,
    - 4, pokud se jedná o buňku na okraji sítě,
    - 3, pokud se jedná o buňku v rohu sítě,
- výsledná matice je pětidiagonální (hlavní diagonála, dvě vedlejší těsně pod a nad hlavní diagonálou, dvě vedlejší diagonály ve vzdálenosti $\pm n$ od hlavní diagonály).

## Úkol 1
**Sestavte matici pomocí formátu `LIL` a vyplňte hodnoty pomocí cyklu přes všechny buňky čtvercové sítě.**
- vypište a vykreslete ji jako obrázek pro $n=3$ a $\varepsilon = 0.1$,
- vykreslete ji jako obrázek pro $n=100$ a $\varepsilon = 0.2$.

````python
import scipy.sparse as sparse
import numpy as np
import matplotlib.pyplot as plt

def A_assembly(n=3, epsilon=1e-1):
    A = sparse.lil_matrix((n*n, n*n))
    for i in range(n):
        for j in range(n):
            aktualni_bunka = i * n + j
            
            pocet_sousedu = 4
            if i == 0: 
                pocet_sousedu = pocet_sousedu - 1     
            if i == n - 1: 
                pocet_sousedu = pocet_sousedu - 1 
            if j == 0: 
                pocet_sousedu = pocet_sousedu - 1     
            if j == n - 1: 
                pocet_sousedu = pocet_sousedu - 1 
            

            A[aktualni_bunka, aktualni_bunka] = 1 + (pocet_sousedu * epsilon)
            
            if i > 0: 
                soused_nahore = (i - 1) * n + j
                A[aktualni_bunka, soused_nahore] = -epsilon
                
            if i < n - 1: 
                soused_dole = (i + 1) * n + j
                A[aktualni_bunka, soused_dole] = -epsilon
                
            if j > 0: 
                soused_vlevo = i * n + (j - 1)
                A[aktualni_bunka, soused_vlevo] = -epsilon
                
            if j < n - 1: 
                soused_vpravo = i * n + (j + 1)
                A[aktualni_bunka, soused_vpravo] = -epsilon
    
    return A

A = A_assembly(3, 1e-1)
print(A.todense())

plt.imshow(A.todense())
plt.colorbar()
plt.title("Matice pro n=3, epsilon=0.1")
plt.show()

A_100 = A_assembly(100, 0.2)
plt.spy(A_100, markersize=0.5) 
plt.title("Nenulové prvky matice pro n=100")
plt.show()
````

- Vykreslete řešení $x$ systému $A \cdot x = b$ jako obrázek čtvercové sítě.

````python
import matplotlib.pyplot as plt

n = 12
epsilon = 2e-1
A = A_assembly(n, epsilon)

b = np.zeros((n, n))
b[n//2 - n//6:n//2 + n//6,n//2 - n//6:n//2 + n//6] = 1
b = b.flatten()
plt.imshow(b.reshape((n, n)))
````

````python
import scipy as sp
import scipy.sparse.linalg as spla

x = spla.spsolve(A, b)
x_jako_mrizka = x.reshape((n, n))
plt.figure(figsize=(6, 5))
plt.imshow(x_jako_mrizka, cmap='plasma')
plt.colorbar()
plt.title("Řešení x systému A * x = b")
plt.show()
````

````python
vlastni_hodnoty, vlastni_vektory = spla.eigsh(A, k=3)


fig, osy = plt.subplots(1, 3, figsize=(15, 4))
fig.suptitle("3 největší vlastní vektory matice A")


for i in range(3):
  
    vektor_mrizka = vlastni_vektory[:, i].reshape((n, n))
    

    graf = osy[i].imshow(vektor_mrizka, cmap='inferno')
    
  
    osy[i].set_title(f"Vektor {i+1}\n(vlastní hodnota: {vlastni_hodnoty[i]:.4f})")
    fig.colorbar(graf, ax=osy[i], fraction=0.046, pad=0.04)

plt.tight_layout()
plt.show()
````
